```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create 6 distributions
BetaDist = np.random.beta(1, 100, 100)

mean = np.mean(BetaDist)
sd = np.std(BetaDist)

plt.hist(BetaDist)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

```python
BinomDist = np.random.binomial(10, 0.25, 100)

mean = np.mean(BinomDist)
sd = np.std(BinomDist)

plt.hist(BinomDist)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

```python
LogisticDist = np.random.logistic(0, 1, 100)

mean = np.mean(LogisticDist)
sd = np.std(LogisticDist)

plt.hist(LogisticDist)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

```python
PowerDist = np.random.power(3, 100)

mean = np.mean(PowerDist)
sd = np.std(PowerDist)

plt.hist(PowerDist)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

```python
PoissonDist = np.random.poisson(10, 100)

mean = np.mean(PoissonDist)
sd = np.std(PoissonDist)

plt.hist(PoissonDist)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

```python
NormalDist = np.random.normal(50, 10, 100)

mean = np.mean(NormalDist)
sd = np.std(NormalDist)

plt.hist(NormalDist)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

```python
norm1 = np.random.normal(5,.5, 100)
norm2 = np.random.normal(10,1, 100)
var3 = norm1 + norm2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```

You can definitely notice the differences between the different sets of distributions. I have taken a few stats courses, so I am familiar with most of these distributions. +/- 1 sd gives the bulk of the data in most cases.
